Построить график на плоскости(разрывы функции в точках не строятся).

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def f(z, a=0, b=0, k=0):
    eps = 0.15
    val = np.sqrt(np.abs(z ** 2 * k ** 3 * 0.25)) / (0.5 + 2 * np.exp(z + 1))
    val += np.power(k, 1/3) / 4 * a * b
    val += np.arctan((z * (z + 3) * (z - 3))/((z ** 2 + 1) * (z - 1)))
    return val

fv = np.vectorize(f)
```

```python
z = np.linspace(-10, 10, 100)
# пометим окрестность точки разрыва
eps = 0.15
z = np.where(np.abs((z - 1)) < eps, np.nan, z)

a = 1
b = 1
k = 1

x = fv(z)

plt.plot(x)
```
