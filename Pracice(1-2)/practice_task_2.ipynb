```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def f(X, Y):
    a = 2 * np.exp(-((X - 5)**2 + (Y + 0.2)**2) / 1.5)
    b = 0.8 * np.exp(-((2*X + 3)**2 + (0.2*Y - 4)**2) / 5)
    c = 4 * np.exp(-((X - 0.2)**2 + (Y + 2)**2) / 2)
    return a + b - c
```

```python
#    x  y
x = np.linspace(-20, 20, 100)
y = np.linspace(-50, 50, 100)
X, Y = np.meshgrid(x, y)

#   z
Z = f(X, Y)

#  
fig = plt.figure(figsize=(12, 8))

#   - 
ax1 = fig.add_subplot(121, projection='3d')
# cmap=plt.cm.PiYG -   ,  - https://matplotlib.org/stable/users/explain/colors/colormaps.html
surf = ax1.plot_surface(X, Y, Z, cmap=plt.cm.PiYG,
                        linewidth=0, antialiased=False)
ax1.set_title('')

#   -  
gs = fig.add_gridspec(2, 2, width_ratios=[1, 1], height_ratios=[0, 3])
ax2 = fig.add_subplot(gs[1, 1])
# ax2 = fig.add_subplot(122)
levels = [-0.99, -0.7, -0.5, -0.3, -0.1, 0.0, 0.1, 0.3, 0.5, 0.7, 0.99]
ct = ax2.contour(X, Y, Z, levels=levels)
cb = plt.colorbar(ct, ax=ax2)
ax2.set_title(' ')

plt.tight_layout()
plt.show()
```
