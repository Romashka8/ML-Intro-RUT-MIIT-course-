### 1. **Данные о прослушивании музыки**  
   - **Регрессия**: По количеству прослушиваний треков разных жанров (рок, электроника, классика) за месяц предсказать, сколько часов пользователь проведёт в музыкальном приложении в следующем месяце.  
   - **Классификация**: Определить, является ли пользователь меломаном (да/нет) на основе разнообразия жанров в его плейлистах.  

### 2. **Данные о передвижении по городу**  
   - **Регрессия**: По истории поездок на такси (время, расстояние, стоимость) предсказать, сколько денег человек потратит на транспорт через неделю.  
   - **Классификация**: Угадать, использует ли человек каршеринг чаще, чем обычное такси (да/нет).  

### 3. **Данные о питании в ресторанах**  
   - **Регрессия**: По среднему чеку и частоте посещений кафе предсказать, сколько раз человек закажет еду на дом в следующем месяце.  
   - **Классификация**: Определить, предпочитает ли человек здоровую пищу (да/нет) на основе выбора блюд (салаты vs фастфуд).
