$\theta_1 = 0.5: J(\theta_1) = \frac{1}{2 * 3} ((0.5 * 1 - 1)^2 + (0.5 * 2 - 2)^2 + (0.5 * 3 - 3)^2) = \frac{7}{12}$

$\theta_1 = 2: J(\theta_1) = \frac{1}{2 * 3} ((2 * 1 - 1)^2 + (2 * 2 - 2)^2 + (2 * 3 - 3)^2) = \frac{7}{3}$

# 5(b)

```python
# 1
xx = np.arange(1, 21)
yy = xx.copy()
```

```python
# a
theta_1 = [-0.5, -0.25, 0.25, 0.5, 1, 1.25]
approx = [xx * theta_1_i for theta_1_i in theta_1]

fig, ax = plt.subplots(2, 3, figsize=(20, 10))

for approx_i, ax_i, theta_1_i in zip(approx, [ax[0][0], ax[0][1], ax[0][2], ax[1][0], ax[1][1], ax[1][2]], theta_1):
    ax_i.scatter(xx, yy, c='indigo')
    ax_i.plot(xx, approx_i, label=f'Approximation line theta_1 = {theta_1_i}', c='deeppink')
    ax_i.legend()
    ax_i.grid(axis='y')
```

```python
# b
def loss(m, x, y, theta):
    def h(x, theta):
        return theta * x

    return 1/(2*m) * np.sum((h(x, theta) - y)**2)

more_thetas = np.linspace(-0.5, 1.25, 250)

losses = [loss(m=more_thetas.shape[0], x=xx, y=yy, theta=theta_1_i) for theta_1_i in more_thetas]

plt.plot(more_thetas, losses, c='deeppink')
min_theta = more_thetas[np.argmin(losses)]
min_value = np.min(losses)
plt.scatter(min_theta, min_value, c='indigo', label=f'min theta = {np.round(min_theta, 2)}')

plt.xlabel('theta')
plt.ylabel('functional')
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
# 2
# np.random.seed(1833)

noise = np.random.uniform(-2, 2, 20)
yy_noise = yy - noise
```

```python
# a
theta_1 = [-0.5, -0.25, 0.25, 0.5, 1, 1.25]
approx = [xx * theta_1_i for theta_1_i in theta_1]

fig, ax = plt.subplots(2, 3, figsize=(20, 10))
for approx_i, ax_i, theta_1_i in zip(approx, [ax[0][0], ax[0][1], ax[0][2], ax[1][0], ax[1][1], ax[1][2]], theta_1):
    ax_i.scatter(xx, yy_noise, c='indigo')
    ax_i.plot(xx, approx_i, label=f'Approximation line theta_1 = {theta_1_i}', c='deeppink')
    ax_i.legend()
    ax_i.grid(axis='y')
```
